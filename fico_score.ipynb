```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)

pd.set_option('display.width', 1000)

#loading the data
df = pd.read_csv('loan_data.csv')
```

```python
# creating buckets
df.info()
```

```python
df['fico_bucket'] = pd.qcut(df['fico_score'], q=8, labels=False)

# creating nice labels
bucket_labels = [
    "350-400 (Bad)",
    "401-550 (Very Poor)",
    "551-600 (Poor)",
    "601-650 (Fair)",
    "651-700 (Good)",
    "701-750 (Very Good)",
    "751-800 (Excelent)",
    "801-850 (Exceptional)"
]
```

```python
display(df.head(10))
```

```python
# maping the number to labels
df['fico_bucket_label'] = pd.qcut(df['fico_score'], q=8, labels=bucket_labels)
```

```python
# checking the results
print(df[['fico_score', 'fico_bucket', 'fico_bucket_label', 'default']].head(10))

print("\nDistribution of buckets:")

print(df['fico_bucket_label'].value_counts().sort_index())
```

```python
# saving
df.to_csv('fico_buckets.csv', index=False)
print("\n✅ File saved as 'fico_buckets.csv'")
```
